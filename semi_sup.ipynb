```python
# semi-supervised topic modeling using anchor words, only 5 words for each topic
# positive: ['great', 'thank', 'help', 'appreci','happi']
# negative: ['delay','late','cancel','rude','no']
# neural: ['flight','airlin','plane','home','destin']
```

```python
import numpy as np
import scipy.sparse as ss
from corextopic import corextopic as ct
```

```python
# read in the matrix data and remove those words with just 1 frequency
import pandas as pd

data = pd.read_csv("matrix_data.csv",header=None)
pd_vocab = pd.read_csv("words_dict.csv",header=None)
idx = data.sum()==1
validcol = [col for i,col in enumerate(data.columns) if not idx[i] ]
validwords = [word for i,word in enumerate(list(pd_vocab[0])) if not idx[i] ]
reduced_data = data[validcol]
#X = ss.csr_matrix(csv)
```

```python
reduced_data
```

```python
X = np.array(reduced_data,dtype=int)
red_df = pd.DataFrame(X)
red_df.to_csv("reduced_data.csv", index=False)
```

```python
X = ss.csr_matrix(X)
```

```python
X.shape
```

```python
TOPIC_NUM = 25
words = np.array(validwords)
topic_model = ct.Corex(n_hidden=TOPIC_NUM,word=words,seed = 6,max_iter=500, verbose=False) 
topic_model.fit(X, words=words);
#topic_model.fit(X, words=words, anchors=[['great', 'thank', 'help', 'appreci','happi'],['delay','late','cancel','rude','no'],['flight','airlin','plane','home','destin']], anchor_strength=4)
```

```python
for n in range(TOPIC_NUM):
    topic_words,_ = zip(*topic_model.get_topics(topic=n))
    print('{}: '.format(n) + ','.join(topic_words))
```

```python
prob_array = topic_model.p_y_given_x
len(prob_array[1])
```

```python
# get the group label for each topic in the document directly from the prob_array
# find the highest value index in array. 
#max([(v,i) for i,v in enumerate(prob_array[1])])[1]
All_true_Topics = [ max([(v,i) for i,v in enumerate(doc_prob)])[1] for doc_prob in prob_array]
All_true_Topics
```

```python
# get the topic sentiment label
topic_groups = []
for n in range(TOPIC_NUM):
    documents_in_each_Topic = [i for i,v in enumerate(All_true_Topics) if v==n]
    topic_groups.append(documents_in_each_Topic)
```

```python
# True Lables  
Origin_data = pd.read_csv("Tweets.csv")
True_label = Origin_data["airline_sentiment"]
```

```python
label_array = []
correct_num = 0
for groups in topic_groups:
    pos_sum = sum([ 1 for value in True_label[groups].values if value == "positive"] )
    neu_sum = sum([ 1 for value in True_label[groups].values if value == "neutral"] )
    neg_sum = sum([ 1 for value in True_label[groups].values if value == "negative"] )
    temp_sum = [pos_sum,neu_sum,neg_sum]
    if pos_sum == max(temp_sum):
        label_array.append("positive")
    if neu_sum == max(temp_sum):
        label_array.append("neutral")
    if neg_sum == max(temp_sum):
        label_array.append("negative")
    correct_num += max(temp_sum)
print(label_array)
print(correct_num / 14640)
```

```python
for n in range(TOPIC_NUM):
    topic_words,_ = zip(*topic_model.get_topics(topic=n))
    print('{}: '.format(n) + ','.join(topic_words) +"------"+ label_array[n])
```

```python
#### get the result matrix 
file_matrix = []
for n in range(len(topic_groups)):
    for group in topic_groups[n]:
        file_matrix.append([group,label_array[n]])   
```

```python
file_label = pd.DataFrame(file_matrix)
file_label
```

```python
tempdf = file_label.sort_values(by=[0])
tempdf.to_csv("corex_result.csv", index=False)
```
